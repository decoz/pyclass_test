##  이미지 관련 

아래는 크기가 100x100 인 0으로 가득찬 2차원 배열을 생성해 이미지로 표현하는 코드입니다. 먼저 이를 실행한 후에 아래의 문제를 풀어보세요. 





````
from PIL import Image
from IPython.display import display
import numpy as np

arr = [] 
for i in range(100):
  tarr = []
  for j in range(100):
    tarr = tarr + [0] 
  arr = arr + [tarr]

display( Image.fromarray( np.array(arr).astype('uint8').reshape(100,100) , 'L' ) )
````

#### 문제 4 : 아래와 같은 이미지가 출력되도록 반복문을 사용하여 배열값을 바꾸세요 [90점]

<img src = 'http://drive.google.com/uc?export=view&id=1nHzjGZqeO49Pk11LUuDOjQpGDH4ESzuS'>

````
# 배열 값 초기화
for x in range(100):
  for y in range(100):
    arr[y][x] = 0 

# 문제 4의 코드를 작성하세요
for x in range(100):
  for y in range(100):
    if x % 5 == 0 :       
      if ( x - y ) * ( 100 - x - y  ) <= 0:
        arr[y][x] = 255
     

    
# 출력
display( Image.fromarray( np.array(arr).astype('uint8').reshape(100,100) , 'L' ) )      
````
