#### 문제 1. 1부터 100 까지의 2나 3의 배수의 합산값을 계산하여 출력하는 프로그램을 작성하세요 [30점]


````python
# 문제 1 코드를 작성하고 a,b,c 값을 바꿔 테스트해보세요
i = 0 
sum = 0 
while i <= 30 :
  if i % 3 == 0 or i % 2 == 0:
    sum += i
  i += 1

print(sum)
````

#### 문제 3. 크기가 3x4 인 임의의 2차원 배열의 값들을 좌측으로 한칸 밀어낸 배열을 출력하세요 [50점]

```
예: 
[4, 1, 3, 6]
[1, 3, 2, 5]
[2, 4, 3, 1]

->

[1, 3, 6, 4]
[3, 2, 5, 1]
[4, 3, 1, 2]
```

````python
# 문제 4 의 코드를 작성하세요
arr = [[4,1,3,6],
       [1,3,2,5],
       [2,4,3,1]]

# 문제 3의 코드를 작성하세요
for y in range(3):
  s = arr[y][0]
  for x in range(3):
    arr[y][x] = arr[y][x+1]
  arr[y][3] = s


for a in arr:
  print(a)

````
