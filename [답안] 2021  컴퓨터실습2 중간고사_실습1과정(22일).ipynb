#### 문제 1. 1부터 30 까지의 숫자를 출력하되 3이나 4의 배수는 '?' 로 표시해 출력하세요 [30점]
```
출력: ? 1 2 ? ? 5 ? 7 ? ? 10 11 ? 13 14 ? ? 17 ? 19 ? ? 22 23 ? 25 26 ? ? 29 ? 
```

````python
# 문제 1 코드를 작성하세요
i = 0 
while i <= 30 :
  if i % 3 == 0 or i % 4 == 0:
    print("?", end = " ")
  else :
    print(i, end = " ")
  i += 1
````

#### 문제 3. 임의의 배열이 있을때 이 배열에서 가장 큰값과 가장 작은값의 위치를 서로 바꿔보세요. 함수는 print함수만 허용됩니다. [50점]

```
예:
[5,3,4,1,2]
->
[1,3,4,5,2]

```

````python

arr = [5,3,7,4,1,2] 

#문제 3 의 코드를 작성하고 배열값을 바꿔며 테스트해보세요
mn,mx = 0,0
for i in range(1,len(arr)):
  if arr[mn] > arr[i] :
    mn = i
  if arr[mx] < arr[i] :
    mx = i

arr[mn], arr[mx] = arr[mx], arr[mn]

# 출력
print(arr)


````
