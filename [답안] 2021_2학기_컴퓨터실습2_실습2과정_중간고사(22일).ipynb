#### 문제 4. 문제 3번을 참조하여 이미지를 x 축 기준 n 픽셀 단위로 영역이 복사되는 blindx(img, n) 함수를 작성하세요.[60점]

```
예: 
blindx(img, 10 )
```
<img src='http://drive.google.com/uc?export=view&id=1RtK1GxmvIScHyl-YO8NHMlz82JZ3ABmH' /><br>



````python
# 이미지를 배열 arr 로 변환
w, h = img.size
arr = np.array(img).reshape(h, w, 3)


def blindx(arr,n):
  for y in range(h) :
    for x in range(w - n) :            
      if x % (2 * n)  < n :
        arr[y][x] = arr[y][x+n].copy()
  return arr

display(Image.fromarray(blindx(arr, 10)))
````
