#### 문제 1. 1부터 n까지의 누적값을 더하는 함수 nsum 을 만드세요 [60점]

예
```
print( nsum(10) )
->
55

```

````python
def nsum(n):
  # 함수의 내용을 작성하세요
  sum = 0
  for i in range(n):
    sum += i + 1
  return sum

  
# 사용 예
print( nsum(10) )
````

#### 문제 2. 문자열에서 시작글자와 끝글자 글자 사이의 문자열을 리턴하는 함수 substr(str, s, e) 를 작성해보세요 [70점]

예
```
s = "abcdef1234"
print(substr(s, 'c', '2' )
->
cdef12
```

시작 글자가 없으면 빈 문자열을, 끝글자가 없으면 문자열 끝까지를 리턴하게 하세요

```
substr(s, 'x', 'z')  -> ''
substr(s, '1', '9')  -> '1234'
```


````python
def substr(str, s, e):
  # 함수의 내용을 작성하세요
  on, result  = 0, ""
  
  for c in str:    
    if c == s :      
      on = 1
    if c == e :
      result += c
      on = 0

    if on == 1 :      
      result += c

  return result 


# 함수 테스트 코드
s = "abcdef1234"
print(substr(s, 'c', '2' ))
  
````

#### 문제 3. 다음 200x200 이 이미지 배열 arr  다음과 같이 출력되도록 작성하세요  [70점]

<img src = 'http://drive.google.com/uc?export=view&id=1_MoywYrbmRqNJH1DlsvvwdKjlJXhBObM' >



````python
# 200 x 200 이미지 배열 생성
arr = np.zeros((200,200))

# 문제의 코드를 작성하세요
for y in range(200):
  for x in range(200):
    arr[y][x] = 255 - 2 * abs(100 - x)

# 이미지로 출력
display( Image.fromarray( np.array(arr).astype('uint8') , 'L' ) )
````

#### 문제 4. 문자와 숫자만으로 이뤄진 임읭의 문자열에서 문자 혹은 숫자가 연결된 갯수들를 출력하는 코드를 작성하세요 [90점]

예:
```
"ab123cc4"
->
2 3 2 1 
```




````python
#문자열을 바꾸면서 테스트해보세요
s = "ab123cc4"

# 문제의 코드를 작성하세요
n,stat = 0, 0
for c in s:    
  if c >= '0' and c <= '9':
    t = 1    
  else :
    t = -1

  if stat * t < 1 :    
    if n > 0 :
      print(n, end = ' ')
    n,stat = 1,t
  else:
    n += 1  

print(n)


````

#### 문제 5. 다음 이미지를 다음과 같이 바꿔보세요 [70점]

이미지 로딩코드를 실행한 후에 


````python
# 이미지 로딩 코드 
url = 'https://modo-phinf.pstatic.net/20180727_74/1532650362069lanVs_JPEG/mosadYIhbW.jpeg?type=round256_256'
img = Image.open(request.urlopen(url) )
display(img)
````

위의 이미지를 이런 형태로 변형하세요

> <img src='http://drive.google.com/uc?export=view&id=1r4qkMYS5wFPbwDGbg950PfzLC7XPQNZn' /><br>


````python
# 이미지 크기를 구한 후에 배열로 변환
w,h  = img.size
arr = np.array(img)[:,:,:3]

# 문제의 코드를 작성하세요
for y in range(h):
  for x in range(w):
    c = arr[y][x]
    r,g,b = c[0],c[1],c[2]    
    
    br =  1 -  abs( w / 2 - x ) * 2 / w 

    r = int( r * br)
    b = int( b * br)
    g = int( g * br)
    
    arr[y][x] = [r,g,b]

# 이미지로 출력
display( Image.fromarray( np.array(arr) ) )
````
